# Using a Tensorflow model to predict capture rate of a pokemon based on its stats:

```python
pok=pd.read_csv(r'C:\Users\jaiva\Jupyter\pokemon.csv',usecols=np.r_[23,19,25,28,35])
X=pok.drop("capture_rate",axis=1)
X=X.values
y=pok["capture_rate"]
y=y.values
y=y.astype(float)
```

```python
def rmse(y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true)))
```

```python
X_train_full, X_test, y_train_full, y_test = train_test_split(X, y)
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full)
```

```python
model = keras.Sequential(
    [
        layers.Dense(2, activation="relu"),
        layers.Dense(3, activation="relu"),
        layers.Dense(4),
    ]
)
model.compile(
    loss=rmse,
    optimizer=Adam(),
    metrics=[rmse]
)
```

```python
history = model.fit(X_train, y_train, epochs=2000, validation_data=(X_valid, y_valid))
```

```python
model.evaluate(X_test, y_test)
```

```python
stats=np.array([[49,49,45,45,49,49,45,45,49,49,45,45,49,49,45,45,49,49]])
new_prediction=model.predict(stats)
print("Prediction: ",new_prediction)
```

```python
parameters = {"n_neighbors": range(1, 50)}
gridsearch = GridSearchCV(KNeighborsRegressor(), parameters)
gridsearch.fit(X_train, y_train)
gridsearch.best_params_
#Used gridsearch so i can find optimal number of n neighbors to use for the model
#I have used 3 in the below cells first just to try
```

```python
train_preds_grid = gridsearch.predict(X_train)
train_mse = mean_squared_error(y_train, train_preds_grid)
train_rmse = sqrt(train_mse)
test_preds_grid = gridsearch.predict(X_test)
test_mse = mean_squared_error(y_test, test_preds_grid)
test_rmse = sqrt(test_mse)
print(train_rmse)
print(test_rmse)
#rmse = Root mean square error
```

```python
#Taking weighted average instead of regular average
parameters = {"n_neighbors": range(1, 50),"weights": ["uniform", "distance"],}
gridsearch = GridSearchCV(KNeighborsRegressor(), parameters)
gridsearch.fit(X_train, y_train)
GridSearchCV(estimator=KNeighborsRegressor(),param_grid={'n_neighbors': range(1, 50),'weights': ['uniform', 'distance']})
print(gridsearch.best_params_)
test_preds_grid = gridsearch.predict(X_test)
test_mse = mean_squared_error(y_test, test_preds_grid)
test_rmse = sqrt(test_mse)
print(test_rmse)
```

**There isn't a significant difference but in some cases weights can reduce the effect of unnecessary outliers making the
model more accurate**
